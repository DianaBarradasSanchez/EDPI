Graficaremos la funci√≥n $f(x')=e^{-\frac{(x'-0.5)^2}{0.01}}$ para resolver:\
a) $x'=x-ct$


```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

x_prime = np.linspace(0, 1, 1000)

y = f(x_prime)

# Para el gr√°fico
plt.figure(figsize=(8, 5))
plt.plot(x_prime, y, 'b-', linewidth=2,color='green')
plt.title('Perturbaci√≥n inicial\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$')
plt.xlabel('Posici√≥n (x\')')
plt.ylabel('Amplitud')
plt.grid(True, linestyle='--',color='purple', alpha=0.7)
plt.xlim(0, 1)
plt.ylim(0, 1.1)


plt.axvline(x=0.5, color='red', linestyle='--', alpha=0.5)
plt.text(0.5, -0.1, 'x\' = 0.5', ha='center', va='top', color='r')

plt.show()
```

b) $x'=x+ct$

```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros
x = np.linspace(0, 1, 500)
c = 1
t = 0.2  # instante de tiempo
sigma = 0.01
x0 = 0.5

# Funci√≥n a visualizar
def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)

# Inciso (a): f(x - ct)
fa = f(x - c*t)

# Inciso (b): f(x + ct)
fb = f(x + c*t)

# Inciso (c): superposici√≥n
fc = fa + fb

# Graficar
fig, axs = plt.subplots(3, 1, figsize=(8, 10),)

axs[0].plot(x, fa,color='green')
axs[0].set_title('Inciso (a): f(x - ct) - desplazamiento a la derecha')
axs[0].grid()

axs[1].plot(x, fb,color='orange')
axs[1].set_title('Inciso (b): f(x + ct) - desplazamiento a la izquierda')
axs[1].grid()

axs[2].plot(x, fc)
axs[2].set_title('Inciso (c): superposici√≥n f(x - ct) + f(x + ct)')
axs[2].grid()

plt.tight_layout()
plt.show()
```
