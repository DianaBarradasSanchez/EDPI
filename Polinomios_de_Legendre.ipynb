```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, simplify, factorial, lambdify, Eq
from sympy.abc import x
from sympy.functions.special.polynomials import legendre
from scipy.special import legendre as scipy_legendre

# Graficar los polinomios de Legendre de n=0 a 5
x_vals = np.linspace(-1, 1, 400)
plt.figure(figsize=(10, 6))
for n in range(6):
    Pn = scipy_legendre(n)
    plt.plot(x_vals, Pn(x_vals), label=f'$P_{{{n}}}(x)$')
plt.title('Polinomios de Legendre $P_n(x)$ para $n = 0, ..., 5$')
plt.xlabel('x')
plt.ylabel('$P_n(x)$')
plt.axhline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Verificar gráficamente la identidad para n=3
n = 3
Pn_expr = legendre(n, x) #expresión de p_3
Pn1_expr = legendre(n + 1, x) #expresión de p_4
lhs = (1 - x**2) * diff(Pn_expr, x)
rhs = (n + 1) * x * Pn_expr - (n + 1) * Pn1_expr
lhs_func = lambdify(x, lhs)
rhs_func = lambdify(x, rhs)

xg = np.linspace(-1, 1, 400)
lhs_vals = lhs_func(xg) #se evaluan los lados de la identidad
rhs_vals = rhs_func(xg)

plt.figure(figsize=(10, 6))
plt.plot(xg, lhs_vals, label='$(1 - x^2)P_3\'(x)$', linestyle='--', color='pink') #Gráfica del polinomio p_3
plt.plot(xg, rhs_vals, label='$(4xP_3(x) - 4P_4(x))$', linestyle=':', color='green',linewidth=3) #Gráfica del polinomio p_4
plt.title('Gráfica de verificación de la identidad para $n = 3$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.axhline(0, color='gray', linewidth=0.5) #Eje x
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Verificación algebraica de la identidad (b)
left_c = simplify((1 - x**2) * diff(Pn_expr, x))
right_c = simplify((n + 1) * x * Pn_expr - (n + 1) * Pn1_expr)
print("\n(c) Identidad (1 - x^2)P_3'(x) = 4xP_3(x) - 4P_4(x):", Eq(left_c, right_c))
```

```python
#Verificación de la identidad de recurrencia
Pn_1_expr = legendre(n - 1, x)
left_d = simplify((n + 1)*Pn1_expr - (2*n + 1)*x*Pn_expr + n*Pn_1_expr)
print("\n(d) Identidad de recurrencia:", Eq(left_d, 0))
```

```python
#Fórmula de Rodrigues
print("\n(e) Verificación de la fórmula de Rodrigues:")
for k in range(6):
    R_expr = (1 / (2*k * factorial(k))) * diff((x**2 - 1)*k, x, k)
    Pk = legendre(k, x)
    correcto = simplify(R_expr - Pk) == 0
    print(f"P_{k}(x) =", R_expr, "==", Pk, "->", correcto)
```
