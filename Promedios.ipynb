Dado que es una distribuciòn uniforme, nuestras hipotesis son:\
$$H_0:\mu= \frac{1}{2}$$\
$$H_a:\mu \neq \frac{1}{2}$$

Calculando los promedios con: $$\bar{r}=\frac{1}{N}\sum_{i=1}^{N}r_i$$\
Donde $r_i$ es cada elemento de la lista y $N$ el total de elementos.\
Asignamos el nivel de significancia de $0.05$



Calculando el estadistico: $$Z_0=\frac{(\bar{r}-\frac{1}{2})\sqrt{N}}{\sqrt{\frac{1}{12}}}$$

```python
def promedios (ri,confianza=0.95):
  n=len(ri)
  media_r=np.mean(ri)
  error_estandar= np.sqrt(1 / 12)
  z0 = (media_r - 0.5)*(np.sqrt(n)) / error_estandar

  alfa = 1 - confianza
  z_critico = stats.norm.ppf(1-alfa/2)
  print(f"Media calculada: {media_r:.5f}")
  print(f"Estadístico Z0: {z0:.5f}")
  print(f"Valor Crítico Z_alfa/2: {z_critico:.5f}")

  if abs(z0) <= z_critico:
          print("Resultado: NO se rechaza H0 ")
  else:
          print("Resultado: Se rechaza H0 ")

promedios(datos)
```
