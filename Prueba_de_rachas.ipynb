Sean las hipótesis:\
$ H_0:$ Los números de la muestra ${x_i}$ son independientes.\
$ H_a:$ Los números de la muestra ${x_i}$ no son independientes.\
Primero se determina $x_i\geq x_{i-1}$ se coloca un signo $+$ o el número $1$; caso contraio,$x_i\leq x_{i-1}$ se coloca un signo $-$ o el número $0$. Posterior a esto se hace el conteo de signos juntos que sean iguales, es decir, las rachas.\
Procedemos con los calculos de $μ$, $\sigma$ y $Z_0$ con las formulas vistas en clase.\
Formaremos un intervalo de aceptacion y con ello sabremos si podemos aceptar o rechazar $H_0$.

```python
import numpy as np

def prueba_rachas(data):
  n = len(data)
  secuencia = []
  for i in range(1, n):
    secuencia.append(1 if data[i] > data[i-1] else 0)

  c0 = 1
  for i in range(1, len(secuencia)):
    if secuencia[i] != secuencia[i-1]:
      c0 += 1

  mu_c0 = (2 * n - 1) / 3
  sigma2_c0 = (16 * n - 29) / 90
  sigma_c0 = np.sqrt(sigma2_c0)
  z0 = (c0 - mu_c0) / sigma_c0
  return c0, mu_c0, sigma2_c0, z0, secuencia

C0, mu, var, Z0, seq = prueba_rachas(datos)

print(f"PRUEBA DE RACHAS")
print(f"Total de datos (n): {len(datos)}")
print(f"Número de rachas (C0): {C0}")
print(f"Media teórica (mu_C0): {mu:.4f}")
print(f"Varianza (sigma2_C0): {var:.4f}")
print(f"Estadístico Z0: {Z0:.4f}")

# Decisión
if abs(Z0) < 1.96:
    print("\nRESULTADO: Se acepta H0. Los números son independientes.")
else:
    print("\nRESULTADO: Se rechaza H0. Los números presentan dependencia.")
```
